```python
import pandas as pd
df=pd.read_excel("/content/drive/MyDrive/A Comparative Study on K-pop and T-pop in the Covid-19 Era/final-term/tpopsongs.xlsx")
df
```

```python
# from pandas.core.arrays.datetimes import DataFrame
import chinese_converter
from cnsenti import Emotion
from cnsenti import Sentiment
emotion = Emotion()
senti = Sentiment()

words_num=[]
good=[]
happy=[]
sad=[]
angry=[]
fear=[]
disgust=[]
shock=[]
positive=[]
negative=[]

for n in range(0,7350) :
  emocnt=emotion.emotion_count(df.iloc[n, 4])
  emo_lst=list(emocnt.values())
  words_num.append(emo_lst[0])
  good.append(emo_lst[2])
  happy.append(emo_lst[3])
  sad.append(emo_lst[4])
  angry.append(emo_lst[5])
  fear.append(emo_lst[6])
  disgust.append(emo_lst[7])
  shock.append(emo_lst[8])

  senticnt = senti.sentiment_calculate(df.iloc[n, 4])
  sent_lst=list(senticnt.values())
  positive.append(sent_lst[2])
  negative.append(sent_lst[3])

df['words_num']=words_num
df['good']=good
df['happy']=happy
df['sad']=sad
df['angry']=angry
df['fear']=fear
df['disgust']=disgust
df['shock']=shock
df['positive']=positive
df['negative']=negative
df
```
