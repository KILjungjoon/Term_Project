```python
sentilist=[]
for n in range(0,7300) :
  df=pd.DataFrame([ko_data.iloc[n,9:-1]/ko_data.iloc[n, -1]])
  sentilist.append(df)

dfsenti=pd.concat(sentilist)
dfsenti
```
