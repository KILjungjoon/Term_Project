```python
# 파일 읽기
import pandas as pd
df=pd.read_excel("/content/drive/MyDrive/A Comparative Study on K-pop and T-pop in the Covid-19 Era/kpopSong.xlsx")
df
```

```python
# 컬럼 'album' 제거
df1=df.drop(labels='album',axis=1)

# 코로나 기간 중 매주 50위 곡 추출
kcovid=df1[(df1['year'] > 2019)&(df1['rank'] < 51)]   # 2020년 이후 50위까지 필터링
kcovid
```

```python
# k_cvd_unq.xlsx를 불러온다.
import pandas as pd
df2=pd.read_excel("/content/drive/MyDrive/A Comparative Study on K-pop and T-pop in the Covid-19 Era/final-term/k_cvd_unq.xlsx")
df2
```

```python
ko_data = pd.merge(left=kcovid, right=df2, how='left')
ko_data
# ko_data.to_excel("/content/drive/MyDrive/A Comparative Study on K-pop and T-pop in the Covid-19 Era/final-term/ko_data.xlsx", index=False)
```

```python
ko_data.describe()
```

* 각 감성 값을 표준화한 데이터프레임 만들기
* 표준화값=각 감성값/tokens_num

```python
# 먼저 데이터프레임 틀 하나를 만든다.
kdata_stnd=ko_data.iloc[:, :9]
kdata_stnd
```

```python
ko_data.columns
```

```python
# 다시 값이 0으로 채워진 빈 컬럼을 생성한다.
kdata_stnd[['sadness', 'joy', 'anger', 'surprise',
       'fear', 'disgust', 'interest', 'boredom', 'pain', 'others', 'neutral',
       'positive', 'negative']]=0
kdata_stnd
```

```python
sentilist=[]
for n in range(0,7300) :
  df=pd.DataFrame([ko_data.iloc[n,9:-1]/ko_data.iloc[n, -1]])
  sentilist.append(df)

dfsenti=pd.concat(sentilist)
dfsenti
```

```python
kdata_stnd=pd.concat([kdata_stnd,dfsenti], axis=1)
kdata_stnd
```
